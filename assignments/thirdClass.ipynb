```python
import numpy as np
import matplotlib.pyplot as plt
# pip install matplotlib
```

```python
fig, ax = plt.subplots(figsize=(3, 2), layout="constrained")
fig.suptitle("Title")

# first axes
ax.set_title("Tabula")

# Show
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(3, 2), layout="constrained")
fig.suptitle("Title")

# first axes
axs[0].set_title("Tabula")
axs[1].set_title("Tabula 2")

# Show
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5), layout="constrained")
ax.plot([2,1,0,3,4,2])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5), layout="constrained")
ax.plot([2,1,0,3,4,2])
ax.plot([4,6,1,3,10,2])

plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y1 = [2, 1, 3, 3, 10]
y2 = [2, 4, 5, 7, 10]

fig, ax = plt.subplots(figsize=(4, 2.5), layout="constrained")
ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5), layout="constrained")
ax.plot(x, y1, label="state 1")
ax.plot(x, y2, label="state 2")

ax.set_title("State of things")
ax.set_xlabel("number")
ax.set_ylabel("state")
ax.legend()

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(4, 2.5), layout="constrained")


axs[0].plot(x, y1, label="state 1")
axs[0].plot(x, y2, label="state 2")

axs[0].set_title("State of things")
axs[0].set_xlabel("number")
axs[0].set_ylabel("state")
axs[0].legend()

axs[1].bar(x, y1, label="state 1")
axs[1].set_title("State of things")
axs[1].set_xlabel("number")
axs[1].set_ylabel("state")
axs[1].legend()

plt.show()
```

```python
dice_throws = np.random.randint(1, 7, 50)
dice_throws
```

```python
dice_throws_number = np.arange(1, 51)
dice_throws_number
```

```python
fig, ax = plt.subplots(figsize=(5,3), layout="constrained")

ax.bar(dice_throws_number, dice_throws)

ax.set_title("Dice throws")
ax.set_xlabel("Throw number")
ax.set_ylabel("Dice value")

plt.show()
```

```python
dice_throws2 = np.random.randint(1, 7, 50)
dice_throws2
```

```python
fig, ax = plt.subplots(figsize=(5,3), layout="constrained")

ax.bar(dice_throws_number[:10] - 0.1, dice_throws[:10], 0.2, label="round 1")
ax.bar(dice_throws_number[:10] + 0.1, dice_throws2[:10], 0.2, label="round 1") # 10 ensimmist valueta, offset jotta ei mene pllekkin, barin paksuus

ax.set_title("Dice throws")
ax.set_xlabel("Throw number")
ax.set_ylabel("Dice value")

ax.set_xlim(0, 50)# tm asettaa leveyden eli voi oikeelle ja vasemmalle asettaa vli 

ax.legend()

plt.show()
```

```python
# Scatterplot, eli palluroita
fig, ax = plt.subplots(figsize=(4,4), layout="constrained")

ax.scatter(dice_throws, dice_throws2)

plt.show()
```
