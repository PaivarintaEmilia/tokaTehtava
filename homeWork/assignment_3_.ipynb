### 3.1.

- Lataa iris datasetti käyttöön

Iris data on 2d array, jossa rivit ovat kukkia (3 erilaista) ja columnit niiden tuntomerkkejä (4 erilaista)



```python
# Otetaan datasetti käyttöön. Seaborn antaa datan valmiissa dataframe muodossa. 
df = sns.load_dataset("iris")
```

```python
# Näytetään kaikki datan eri osat, sepal_length, sepal_width, petal_length, petal_width, species
df.head()
#df.info()
```

- Luodaan datasta histogrammi käyttämällä matplotlib työkalua
- Histogrammi näyttää jonkun datan jakauman, eli montako arvoa on tietyllä välillä
- Otetaan tarkasteluun sepal_width ja Setosa species. Kuinka monta setosaa sai tietyn sepal_width arvon.

```python
# Otetaan tarkasteluun Setosa-lajikkeen omat rivit
setosa = df[df["species"] == "setosa"]
setosa
```

```python
# Luodaan histogrammi, jossa tarkasteluun otetaan sepal_width
plt.hist(setosa["sepal_width"], bins=10, color='lightblue', edgecolor='grey')

plt.title("Sepal widths of Setosa")
plt.xlabel("Sepal width")
plt.ylabel("Cuantity")
plt.show()
```

## 3.3. Luodaan horisontaalinen pairplot Seabornin avulla

- Datana käytetään data sourcea, joka kertoo, kuinka monta ulkomaalaista tai suomalaista on yöpynyt Lapissa

Data tuotiin ohjeiden linkin kautta. Se sisältää vuodet, monta yötä on vietetty lapissa sekä minkä maalaisia vierailijat ovat olleet. Data on JSON muodossa ja se otetaan käyttöön Pandas-kirjaston avulla. 

```python
df_lapland_visits = pd.read_excel("../laplandVIsits.xlsx")



df_lapland_visits.head()
```

Tulee määrittää x ja y akselit. Y = vuosi ja X = yövytyt yöt

https://seaborn.pydata.org/examples/part_whole_bars.html

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 12), layout="constrained")

# Muutetaan tyyppi, jotta horisontaalinen
df_lapland_visits["Years"] = df_lapland_visits["Years"].astype(str)


# Suodatetaan vain Finland-rivit
df_finland = df_lapland_visits[df_lapland_visits["Country"] == "Finland"]
sns.set_color_codes("pastel")
# Plotataan Finland
sns.barplot(
    x="Nights", 
    y="Years", 
    data=df_finland,
    label="Finland", 
    color="b"
)

# Suodatetaan vain Foreign-rivit
df_foreign = df_lapland_visits[df_lapland_visits["Country"] == "Foreign"]
sns.set_color_codes("muted")
# Plotataan foreign countries
sns.barplot(
    x="Nights", 
    y="Years", 
    data=df_foreign,
    label="Foreign Countries", 
    color="b"
)




# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
```
